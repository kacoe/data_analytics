```
import pandas as pd

dados1 = pd.read_excel("/content/drive/MyDrive/Acs/dados1_ac02.xlsx", header=None)
serie_completa = dados1.stack().reset_index(drop=True)


serie_completa.to_excel("/content/drive/MyDrive/Acs/series.xlsx")
```

```
import numpy as np
from statistics import mode

#1.2
media = np.mean(serie_completa)
moda = mode(serie_completa)
mediana = np.median(serie_completa)

#1.3
variancia = np.var(serie_completa)
desvio_padrao = np.std(serie_completa)
quartis = serie_completa.quantile([0.25, 0.5, 0.75])
variancias_arredondadas = variancia.round(2)
desvios_padrao_arredondados = desvio_padrao.round(2)

dados_resumo = pd.DataFrame({
    'Média': [media],
    'Moda': [moda],
    'Mediana': [mediana],
    'Variância': [variancias_arredondadas],
    'Desvio Padrão': [desvios_padrao_arredondados],
    'Primeiro Quartil': [quartis[0.25]],
    'Segundo Quartil ': [quartis[0.5]],
    'Terceiro Quartil': [quartis[0.75]]
})

dados_resumo.to_excel("/content/drive/MyDrive/Acs/resultados_pt1.xlsx")

```
