```python
import pandas as pd

skip = 8
dados2 = pd.read_excel("/content/drive/MyDrive/Acs/dados2_ac02.xlsx", skiprows=skip)

dados2.to_excel("/content/drive/MyDrive/Acs/dados.xlsx")
```

```python
import numpy as np
from statistics import mode

colunas_especificas = dados2.iloc[:, [ 1, 2, 3, 4]]

medias = colunas_especificas.mean()
modas = colunas_especificas.mode().iloc[0]  # Pega a primeira moda, pode haver mais de uma
medianas = colunas_especificas.median()

resultados = pd.DataFrame({
    "Médias": medias,
    "Modas": modas,
    "Medianas": medianas,
})

resultados.to_excel("/content/drive/MyDrive/Acs/resultados_pt2.xlsx")

```

```python
import numpy as np
from statistics import mode

colunas_especificas = dados2.iloc[:, [ 1, 2, 3, 4]]

variancias = colunas_especificas.var()
desvios_padrao = colunas_especificas.std()
quartis = colunas_especificas.quantile([0.25, 0.50, 0.75])
variancias_arredondadas = variancias.round(2)
desvios_padrao_arredondados = desvios_padrao.round(2)


resultados = pd.DataFrame({
    "Variâncias": variancias_arredondadas,
    "Desvios Padrão": desvios_padrao_arredondados,
    "Quartis 25%": quartis.loc[0.25],
    "Quartis 50%": quartis.loc[0.50],
    "Quartis 75%": quartis.loc[0.75]
})
resultados.to_excel("/content/drive/MyDrive/Acs/resultados2_pt2.xlsx")
```
